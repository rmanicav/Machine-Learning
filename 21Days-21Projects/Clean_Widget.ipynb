```python
import json

input_nb = "/content/1_Data_Storytelling_Analysing_Survival_on_the_Titanic.ipynb"
output_nb = "/content/1_Data_Storytelling_Analysing_Survival_on_the_Titanic_clean.ipynb"

with open(input_nb, "r", encoding="utf-8") as f:
    nb = json.load(f)

# Remove top-level widgets metadata
nb.get("metadata", {}).pop("widgets", None)

# Remove widget metadata from all cell outputs
for cell in nb.get("cells", []):
    if "outputs" in cell:
        for output in cell["outputs"]:
            if "metadata" in output and "widgets" in output["metadata"]:
                output["metadata"].pop("widgets", None)

with open(output_nb, "w", encoding="utf-8") as f:
    json.dump(nb, f, indent=1)

print("Clean notebook saved as:", output_nb)
```
